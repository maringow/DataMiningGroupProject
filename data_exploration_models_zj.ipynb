## Random Forest Regression

For now, I'm going to try and exclude the outcomes data.

```python
#create random forest regression function
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=0)
```

```python
#Trim data set to features to be utilized in model
dataset = dataset.loc[:,["PSI_4_SURG_COMP", "H_CLEAN_LINEAR_SCORE", "H_COMP_1_LINEAR_SCORE", "H_COMP_2_LINEAR_SCORE",\
              "H_COMP_3_LINEAR_SCORE", "H_COMP_5_LINEAR_SCORE", "H_COMP_6_LINEAR_SCORE", \
              "H_COMP_7_LINEAR_SCORE", "H_HSP_RATING_LINEAR_SCORE", "H_QUIET_LINEAR_SCORE", "H_RECMND_LINEAR_SCORE",\
              "ZIP Code", "Number of Completed Surveys", "Survey Response Rate Percent"]].reindex()
```

```python
#Convert objects into categorical
#dataset["State"] = dataset["State"].astype("category")
```

```python
#from sklearn.preprocessing import LabelEncoder
#label_encoder = LabelEncoder()
```

```python
#Define x and y
x = dataset.iloc[:,1:]
y = dataset.iloc[:,0]
```

```python
#Define train_x and train_y
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Fit the model
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
```

### Evaluate Model

```python
#look at y_test vs. y_pred
plt.scatter(x = y_test, y = y_pred)
```

```python
#look at residuals
resids = y_test - y_pred

plt.scatter(x = range(x_test.shape[0]), y = resids)
```

```python
# Create a pd.Series of features importances
importances_rf = pd.Series(rf.feature_importances_,
index = x.columns)
# Sort importances_rf
sorted_importances_rf = importances_rf.sort_values()
# Make a horizontal bar plot
sorted_importances_rf.plot(kind='barh', color='blue')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error as MSE
```

```python
#test RMSE
# Compute mse_dt
mse_dt = MSE(y_test, y_pred)

# Compute rmse_dt
rmse_dt = mse_dt**(1/2)

# Print rmse_dt
print("Test RMSE: {:.5f}".format(rmse_dt))
```
